```python
import pyreadstat
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
```

```python
df = {}
df['Hours'] = [23, 12, 15, 14, 16, 21, 14, 11, 18, 9]
df['GPA'] = [3.95, 3.90, 4.00, 3.46, 3.97, 3.89, 3.66, 3.91, 3.80, 3.89]
```

```python
r = pearsonr(df['Hours'], df['GPA']).statistic
print(f'pearson: {r}')
r_d = r ** 2
print(f'determination: {r_d}')
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x=df['Hours'], y=df['GPA'])
```
