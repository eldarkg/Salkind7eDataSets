```python
import pyreadstat
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
```

```python
df,_ = pyreadstat.read_sav('datasets/Chapter 5 Data Set 5.sav')
```

```python
r = pearsonr(df['Exercise'], df['GPA']).statistic
print(f'pearson: {r}')
r_d = r ** 2
print(f'determination: {r_d}')
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x=df['Exercise'], y=df['GPA'])
```
